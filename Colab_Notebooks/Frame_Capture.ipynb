Program To read video and extract Frames.

```python
import cv2


# Function to extract frames
def FrameCapture(path):
      
    # Path to video file
    vidObj = cv2.VideoCapture(path)
  
    # Used as counter variable for frames
    count = 0
    
    #Used for naming image
    #img_num = 0
  
    # checks whether frames were extracted
    success = 1
  
    while success:
  
        # vidObj object calls read
        # function extract frames
        success, image = vidObj.read()
        # Saves the 5th frame with frame-count
        if count%7 == 0:
            cv2.imwrite("drive/MyDrive/Terrain_Classification/Train_Dataset/Road/frame%d.jpg" % int(count/7), image)
        count += 1
        
    vidObj.release()  

# Driver Code
if __name__ == '__main__':
  
    # Calling the function
    FrameCapture("drive/MyDrive/Terrain_Classification/Train_Videos/Road_1.mp4")
```
